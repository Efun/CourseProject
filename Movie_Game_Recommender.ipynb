```python
movies_cleaned = movies[["genres", "keywords", "title", "overview", "production_companies", "tagline"]]
```

```python
movies_cleaned.iloc[0]
```

```python
movies_parsed = movies_cleaned[["title", "overview", "tagline"]]
movies_parsed['genres'] = movies_cleaned['genres'].map(lambda x: ",".join(map(lambda y: y['name'], json.loads(x))))
movies_parsed['keywords'] = movies_cleaned['keywords'].map(lambda x: ",".join(map(lambda y: y['name'], json.loads(x))))
movies_parsed['production_companies'] = movies_cleaned['production_companies'].map(lambda x: ",".join(map(lambda y: y['name'], json.loads(x))))
movies_parsed = movies_parsed.dropna(subset=['overview', 'keywords', 'title'])
```

```python
movies_parsed.to_csv('movies_parsed.csv')
steam_parsed.to_csv('steam_parsed.csv')
```

```python
import numpy as np
import torch
import os
import pandas as pd
import faiss
import time
from sentence_transformers import SentenceTransformer, util
```

```python
movies = pd.read_csv('movies_parsed.csv')
steam = pd.read_csv('steam_parsed.csv')
```

```python
steam.head()
```

```python
movies.head()
```

```python
movies_data = movies['overview'].to_list()
steam_data = steam['desc_snippet'].to_list()
```

```python
# model = SentenceTransformer('msmarco-distilbert-base-v4')
model = SentenceTransformer('all-mpnet-base-v2')
# model = SentenceTransformer()
encoded_movies_data = model.encode(movies_data)
encoded_steam_data = model.encode(steam_data)
```

```python
encoded_movies_data.shape
```

```python
movies_index = faiss.IndexIDMap(faiss.IndexFlatIP(768))
movies_index.add_with_ids(encoded_movies_data, np.array(range(0, len(movies_data))))

steam_index = faiss.IndexIDMap(faiss.IndexFlatIP(768))
steam_index.add_with_ids(encoded_steam_data, np.array(range(0, len(steam_data))))
```

```python
faiss.write_index(movies_index, 'movies_index_symmetric')
faiss.write_index(steam_index, 'steam_index_symmetric')
```

```python
movies_index = faiss.read_index('movies_index')
steam_index = faiss.read_index('steam_index')
```

```python
res = faiss.StandardGpuResources()
gpu_movies_index = faiss.index_cpu_to_gpu(res, 0, movies_index)
gpu_steam_index = faiss.index_cpu_to_gpu(res, 0, steam_index)
```

```python
print(movies.iloc[0])
print(movies.iloc[0].overview)
query = model.encode([movies.iloc[0].overview])
top_k = gpu_steam_index.search(query, 10)
print([steam.iloc[_id]['name'] for _id in top_k[1].tolist()[0]])
```

```python
print(steam.iloc[1])
print(steam.iloc[1].desc_snippet)
query = model.encode([steam.iloc[1].desc_snippet])
top_k = gpu_movies_index.search(query, 10)
print([movies.iloc[_id]['title'] for _id in top_k[1].tolist()[0]])
```
